```python
plt.figure(figsize=(15,10))
plt.subplot(2,3,1)
plt.scatter(x=cars93['Length'],y=cars93['Width'],color='r')
plt.xlabel('length of car')
plt.ylabel('width of car')
plt.title('length vs width')
plt.subplot(2,3,2)
plt.scatter(x=cars93["EngineSize"],y=cars93['Horsepower'],color='#f0932b')
plt.xlabel('Engine size')
plt.ylabel('Horsepower')
plt.title('Engine Vs Horsepower')
plt.subplot(2,3,3)
plt.scatter(x=cars93['Min.Price'],y=cars93["Max.Price"],color='#eb4d4b')
plt.xlabel("Minimum price")
plt.ylabel("Maximum price")
plt.title("Minimum price vs Maximum price")
plt.subplot(2,3,4)
plt.scatter(x=cars93['Wheelbase'],y=cars93["Turn.circle"],color='#40739e')
plt.xlabel('wheel base')
plt.ylabel('Turn Circle')
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,3,1)
plt.hist(x=cars93['Length'],color='#eb4d4b')
plt.xlabel('Length of car')
plt.ylabel('Frequecy')
plt.subplot(2,3,2)
plt.hist(x=cars93['Width'],color='#badc58')
plt.xlabel('Width of car')
plt.ylabel('Frequecy')
plt.subplot(2,3,3)
plt.hist(x=cars93['Wheelbase'],color='#ff7979')
plt.xlabel('Wheelbase of car')
plt.ylabel('Frequecy')
plt.subplot(2,3,4)
plt.hist(x=cars93['Weight'],color='#6ab04c')
plt.xlabel('Weight of car')
plt.ylabel('Frequecy')
plt.subplot(2,3,5)
plt.hist(x=cars93['EngineSize'],color='#7ed6df')
plt.xlabel('Engine Size of car')
plt.ylabel('Frequency')
plt.subplot(2,3,6)
plt.hist(x=cars93['Price'],color='#30336b')
plt.xlabel('price per head passenger')
plt.ylabel('Frequency')
```
