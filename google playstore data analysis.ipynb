```python
plt.figure(figsize=(20,8))
sns.barplot(x=category_count.index,y=category_count.values)
plt.xlabel('Types of Apps')
plt.ylabel('Number of Categorical Apps')
plt.xticks(rotation=90)
plt.title('Number of Categorical Apps vs Types of Apps')
plt.grid(True)
```
